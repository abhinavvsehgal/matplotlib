```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np
```

```python

fig = plt.figure()
ax = plt.axes(projection = "3d")

z = np.linspace(0,15, 1500)
x = np.cos(z)
y = np.sin(z)
ax.plot3D(x, y, z, 'gray')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection = "3d")

z = np.linspace(0,15, 1500)
x = np.cos(z)
y = np.sin(z)
ax.plot3D(x, y, z, 'gray')



zp = 15 * np.random.random(100)
xp = np.cos(zp) + 0.1 * np.random.randn(100)
yp = np.sin(zp) + 0.1 * np.random.randn(100)
ax.scatter3D(xp,yp,zp, c= zp, cmap = 'hsv')
plt.show()
```
