# Movie Data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Downloads/machine-learning-june-2019-master/Dataset/movie_metadata.csv")
df

```

```python
dirName = np.array(df['director_name'])[:20]
dirName
```

```python
dirLikes= np.array(df['director_facebook_likes'])[:20]
dirLikes
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (30,10))
plt.bar(dirName, dirLikes, width = 0.5, color = 'green',label = '2018')
plt.xlabel('Director Name')
plt.ylabel('Facebook Likes')
plt.show()
```

```python
critics = np.array(df['num_critic_for_reviews'])[:20]
critics
```

```python
movieFbLikes = np.array(df['movie_facebook_likes'])[:20]
movieFbLikes
```

```python
plt.figure(figsize = (30,10))
plt.bar(dirName, movieFbLikes, width = 0.3, color = 'blue')
plt.show()
```

```python
rating = np.array(df['imdb_score'])[:20]
rating
```

```python
plt.figure(figsize = (30,10))
plt.bar(dirName, rating, width = 0.4, color = 'violet')
plt.xlabel('Directors')
plt.ylabel('Ratings')
plt.title('Directors vs IMDB rating')

plt.show()

```

# Public place analysis

```python
Data = pd.read_csv("Downloads/analysis-public-place-assaults-sexual-assaults-and-robberies-2015-csv.csv")
Data
```

```python
Data.shape
```

```python
area = np.array(Data['Area_unit_2013_label'])[:20]
area
vict = np.array(Data['Victimisations_calendar_year_2015'])[:20]
vict
```

```python
pop = np.array(Data['Population_mid_point_2015'])[:20]
pop
```

```python
plt.figure(figsize = (51,20))
plt.bar(area, pop, width = 0.3, color = 'grey', label = '2015')
plt.bar(area, vict, width = 0.3, color = 'black', label = ' 2016')
plt.xlabel('places')
plt.ylabel('number of assaults')
plt.legend()
plt.show()
```
